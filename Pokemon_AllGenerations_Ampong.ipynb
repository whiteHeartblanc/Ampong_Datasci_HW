###  Import requests library


```python
import requests
import json
URL="https://bulbapedia.bulbagarden.net/wiki/List_of_Pok%C3%A9mon_by_National_Pok%C3%A9dex_number#List_of_Pok.C3.A9mon_by_National_Pok.C3.A9dex_number"
page=requests.get(URL)
```

### Load the page

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(page.content, 'html.parser')
```

```python

poke_content=soup.find(id='mw-content-text')

poke_tables=poke_content.find_all('table')
```

## Checking and getting the first 3 pokemon for each generation details 
### This is to check if reading all the pokemon works and first 3 so that its faster to check and load


```python
urlpoke='https://bulbapedia.bulbagarden.net/'
for j in range (1, 9):
    currentgen= poke_tables[j]
    info_start = 3
    dextype= currentgen.contents[1].contents[1].text.strip()
    gen = 'Generation ' + str(j) 
    print(f'{gen} {dextype}')
    for i in range(info_start, 8, 2):
        poke_info=currentgen.contents[i]
        kdex=poke_info.contents[1].text.strip()
        ndex=poke_info.contents[3].text.strip()
        name=poke_info.contents[7].text.strip()
        pokeurl= urlpoke+poke_info.contents[7].find("a")["href"]
        type1=poke_info.contents[9].text.strip()
        if len(poke_info.contents) > 10:
            type2=poke_info.contents[11].text.strip()
            print(f'Pokemon {kdex} {ndex} {name} find more at {pokeurl} is a {type1} & {type2} Pokemon')
        else:
            print(f'Pokemon {kdex} {ndex} {name} find more at {pokeurl} is a {type1} Pokemon')
```

## Now applying the same loop but for all and saving in JSON file

```python


for j in range (1, 9):
    currentgen= poke_tables[j]
    info_start = 3
    dextype= currentgen.contents[1].contents[1].text.strip()
    gen_json_name = 'Pokemon Generation ' + str(j) + '.json'
   # print(f'Gen {j} {dextype}')
    gen_json = []
    for i in range(info_start,  len(currentgen.contents), 2):
        poke_info=currentgen.contents[i]
        dex=poke_info.contents[1].text.strip()
        ndex=poke_info.contents[3].text.strip()
        name=poke_info.contents[7].text.strip()
        type1=poke_info.contents[9].text.strip()
        pokeurl= urlpoke+poke_info.contents[7].find("a")["href"]
        if len(poke_info.contents) > 10:
            type2=poke_info.contents[11].text.strip()
            gen_json.append({
                dextype: dex,
                "ndex": ndex,
                "name": name,
                "type1": type1,
                "type2": type2,
                "pokemonurl": pokeurl
            })
        else:
            gen_json.append({
                dextype: dex,
                "ndex": ndex,
                "name": name,
                "type1": type1,
                "pokemonurl": pokeurl
            })
    gen_json
    print(gen_json)
    with open(gen_json_name, 'w') as outfile:
        json.dump(gen_json, outfile, indent=4)
    
```
