## Chose MTV news getting dates from March 11-12 2021
### I chose MTV because some news websites had different format and the like so the from parameter was not working correctly for some websites

```python



news= "https://newsapi.org/v2/everything?"
parameters = {
        'sources' : 'mtv-news',
        'from' : '2021-03-11',
       'to': '2021-03-12',
     'apikey' : '38380371a50140beae979b6e6d5e2a32',
       
        
}


req = requests.get(news, parameters)
newsJson= req.json()
newsJson['articles']
```

### Outing the JSON file 

```python
with open('mtvnews.json', 'w') as outfile:
  json.dump(newsJson['articles'], outfile, indent=4)
```
